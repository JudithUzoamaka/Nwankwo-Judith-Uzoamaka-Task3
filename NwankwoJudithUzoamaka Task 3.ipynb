3) To load the CSV Files

```python
import pandas as pd

orders = pd.read_csv("pizza_sales/pizza_sales/orders.csv")
order_details = pd.read_csv("pizza_sales/pizza_sales/order_details.csv")
pizzas = pd.read_csv("pizza_sales/pizza_sales/pizzas.csv")
pizza_types = pd.read_csv("pizza_sales/pizza_sales/pizza_types.csv", encoding="latin1")
```

4) To merge All Tables

```python

sales_df = (
    order_details
    .merge(orders, on="order_id")
    .merge(pizzas, on="pizza_id")
    .merge(pizza_types, on="pizza_type_id")
)

sales_df.head()
```

5) To Create Revenue Column

```python
sales_df["revenue"] = sales_df["quantity"] * sales_df["price"]
```

6) Total Revenue

```python
sales_df["revenue"].sum()
```

7) Total Quantity Sold

```python
sales_df["quantity"].sum()
```

8)Total Orders

```python
orders["order_id"].nunique()
```

9)Number of Pizza Types

```python
pizza_types["pizza_type_id"].nunique()
```

10) Average Price

```python
pizzas["price"].mean()
```

11) Peak Hours

```python
sales_df["order_hour"] = pd.to_datetime(sales_df["time"]).dt.hour

sales_df.groupby("order_hour")["revenue"].sum().sort_values(ascending=False)
```

12)Sales by Day of Week

```python
sales_df["order_date"] = pd.to_datetime(sales_df["date"])
sales_df["day"] = sales_df["order_date"].dt.day_name()

sales_df.groupby("day")["revenue"].sum().sort_values(ascending=False)
```

13)Top 5 Bestselling Pizzas

```python
sales_df.groupby("name")["quantity"].sum().sort_values(ascending=False).head(5)
```

16) Monthly Sales

```python
sales_df["month"] = sales_df["order_date"].dt.month_name()

sales_df.groupby("month")["revenue"].sum()
```

17)Least Selling Pizzas

```python
sales_df.groupby("name")["quantity"].sum().sort_values().head(5)
```
